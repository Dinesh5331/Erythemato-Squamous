```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
df1 = pd.read_csv('train_dataset.csv')
df1
```

```python
X_train_sm = df1.drop('class', axis=1)
y_train_sm = df1['class']
```

```python
import pandas as pd
df2 = pd.read_csv('test_dataset.csv')
df2
```

```python
X_test = df2.drop('class',axis=1)
y_test = df2['class']
```

```python
# Random Forest training and evaluation
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1) Create & train the model
rf = RandomForestClassifier(
    n_estimators=200, 
    random_state=42,
)

rf.fit(X_train_sm, y_train_sm)

# 2) Predict on the test set
y_pred_rf = rf.predict(X_test)

# 3) Metrics
acc_rf = accuracy_score(y_test, y_pred_rf)
print("Random Forest Accuracy:", acc_rf)
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_rf))
```

```python
cm_rf = confusion_matrix(y_test, y_pred_rf)
# 4) Optional: confusion matrix heatmap
plt.figure(figsize=(7,6))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Random Forest Confusion Matrix")
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier

# --- Base estimator ---
rf = RandomForestClassifier(random_state=42)
  
# --- Hyperparameter grid (balanced between coverage and runtime) ---
param_grid_rf = {
    "n_estimators": [200, 400],                # more trees for stability
    "max_depth": [None, 20, 40],               # allow deeper trees to capture complexity
    "max_features": ["sqrt", "log2", 0.5],     # try fraction of features as well
    "min_samples_leaf": [1, 2, 4],             # smooth small leaves
    "min_samples_split": [2, 5],               # control minimum split size
    "class_weight": [None, "balanced"],        # account for any residual imbalance
    "bootstrap": [True]                        # keep bootstrap (typical for RF)
}

# --- Stratified K-Fold ---
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# --- GridSearchCV ---
grid_rf = GridSearchCV(
    estimator=rf,
    param_grid=param_grid_rf,
    scoring="f1_macro",   # good for multiclass imbalance
    n_jobs=-1,
    cv=cv,
    verbose=2,
    refit=True
)

# --- Fit on SMOTE-balanced training data (expects X_train_sm, y_train_sm in scope) ---
grid_rf.fit(X_train_sm, y_train_sm)

# --- Print best parameters and best CV score ---
print("Best Parameters (Random Forest):", grid_rf.best_params_)
print("Best CV Score (f1_macro):", grid_rf.best_score_)
```

```python
# Train & test Random Forest using best parameters from grid_rf
import joblib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Use the refitted best estimator from GridSearchCV (grid_rf must be in scope)
best_rf = grid_rf.best_estimator_

# If for some reason best_rf isn't fitted (grid_rf.refit=True normally fits it already),
# fitting again is harmless and ensures it's trained on the full SMOTE training set:
best_rf.fit(X_train_sm, y_train_sm)

# Predict on test set
y_pred_rf_best = best_rf.predict(X_test)

# Accuracy
acc = accuracy_score(y_test, y_pred_rf_best)
print(f"Random Forest (tuned) Test Accuracy: {acc}\n")

# Classification report
print("Classification Report:\n")
print(classification_report(y_test, y_pred_rf_best))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred_rf_best)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Random Forest (Tuned) — Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# AUC + ROC (One-vs-Rest) for tuned Random Forest
from sklearn.metrics import roc_curve, auc, roc_auc_score
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
import numpy as np

# 1) Predict probabilities (RandomForest supports predict_proba)
y_score_rf = best_rf.predict_proba(X_test)

# 2) Binarize y_test for one-vs-rest ROC
classes = np.unique(y_test)
y_test_bin = label_binarize(y_test, classes=classes)

# 3) Macro AUC (multi_class='ovr')
auc_macro_rf = roc_auc_score(y_test_bin, y_score_rf, multi_class='ovr', average='macro')
print("Random Forest Macro AUC (ovr, macro):", auc_macro_rf)

per_class_auc = {}
for i, cls in enumerate(classes):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_score_rf[:, i])
    per_class_auc[cls] = auc(fpr, tpr)   # store AUC for each class

# 5) Plot ROC curves (one line per class)
plt.figure(figsize=(6, 4))
for i, cls in enumerate(classes):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_score_rf[:, i])
    plt.plot(fpr, tpr, label=f"Class {cls} (AUC = {per_class_auc[cls]:.3f})")

# Diagonal
plt.plot([0, 1], [0, 1], linestyle='--', linewidth=1)

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Random Forest ROC Curves (One-vs-Rest)")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
importances = best_rf.feature_importances_
feature_names = [f"Feature {i}" for i in range(X_train_sm.shape[1])]

# Create DataFrame
importance_df = pd.DataFrame({
    "Feature": feature_names,
    "Importance": importances
}).sort_values(by="Importance", ascending=False)

print(importance_df)
```

```python
# Plot feature importance
plt.figure(figsize=(10,6))
plt.barh(importance_df["Feature"], importance_df["Importance"])
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.title("Random Forest Feature Importance (UCI Dermatology)")
plt.gca().invert_yaxis()  # highest importance at top
plt.show()
```
